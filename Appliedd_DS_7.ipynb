````python
# create a function which will validate the password ?

def validate(password):
  is_upper = False
  is_lower = False
  is_digit = False
  is_symbol = False

  if len(password) <20 and len(password)> 7:
    for char in password:
      if char.isupper():
        is_upper = True
      elif char.islower():
        is_lower = True
      elif char.isdigit():
        is_digit = True
      elif char in ('@', '!','#','$','%'):
        is_symbol = True
    if is_upper and is_lower and is_digit and is_symbol:
      print("password is valid")
    else:
      print("password is not valid")
  else:
    print("password is not valid")
````

````python
# create a function which will help you find outliers from the column
def outlier(col):
    quat1 = df[col].quantile(.25)
    quat3 = df[col].quantile(.75)

    iqr = quat3 - quat1
    upper_limit = quat3 + 1.5 * iqr
    lower_limit = quat1 - 1.5 * iqr

    return df[(df[col] > upper_limit) | (df[col] < lower_limit)]
````

````python
outlier('age')
````

````python
def outlier_columns(column_name):
  q1 = df[column_name].quantile(0.25)
  q3 = df[column_name].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  return lower_bound, upper_bound
````

````python
def build_model(hp):
  model = Sequential()

  model.add(Flatten(input_shape = (28,28))), # input layer

  for i in range(hp.Int('num_layers', min_value = 1 , max_value = 4)):
    model.add(Dense(units = hp.Int('units_' +str(i) , min_value = 1  , max_value = 128 , step = 20),
                    activation = hp.Choice("act_" +str(i), ['relu', 'tanh']))),


  model.add(Dense(10 , activation = 'softmax'))

  model.compile(loss = 'sparse_categorical_crossentropy', optimizer = hp.Choice('optimizer', ['adam','rmsprop'] ), metrics = ['accuracy'])

  return model
````

````python
def top_value(data , num):
  name = []
  count = 0
  for i in ast.literal_eval(data):
    if count <= num:
      name.append(i['name'])
      count += 1
    else:
      break
  return name
````

````python
def process_text(tag):
  ps = PorterStemmer()


  txt = re.sub(r"[^\w\s]", "", tag)

  txt = word_tokenize(txt)


  txt = [ps.stem(word) for word in txt if word not in stopwords.words('english')]

  txt = " ".join(txt)
  return txt
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator , TransformerMixin
````

````python
class preprocess(BaseEstimator , TransformerMixin):
  def fit(self , X, y=None ):
    self.jobencoder= LabelEncoder()
    self.jobencoder.fit(X['job'])
    self.maritalencoder = LabelEncoder()
    self.maritalencoder.fit(X['marital'])
    self.educationencoder = LabelEncoder()
    self.educationencoder.fit(X['education'])
    self.defaultencoder = LabelEncoder()
    self.defaultencoder.fit(X['default'])
    self.housingencoder = LabelEncoder()
    self.housingencoder.fit(X['housing'])
    self.loanencoder = LabelEncoder()
    self.loanencoder.fit(X['loan'])
    self.monthencoder = LabelEncoder()
    self.monthencoder.fit(X['month'])
    return self

  def transform(self, X, y = None):
    X['job'] = self.jobencoder.transform(X['job'])
    X['marital'] = self.maritalencoder.transform(X['marital'])
    X['education'] = self.educationencoder.transform(X['education'])
    X['default'] = self.defaultencoder.transform(X['default'])
    X['housing'] = self.housingencoder.transform(X['housing'])
    X['loan'] = self.loanencoder.transform(X['loan'])
    X['month'] = self.monthencoder.transform(X['month'])

    return X
````

````python
from sklearn.pipeline import Pipeline
numeric_pipe = Pipeline([('scaler', StandardScaler())])
categorical_pipe = Pipeline([('encodiing', preprocess())])
````

````python
X.select_dtypes(include = 'object').columns
````

````python
transform = ColumnTransformer([('num', numeric_pipe, ['age', 'balance', 'day_of_week', 'duration', 'campaign', 'pdays', 'previous']),
                              ('cat', categorical_pipe , ['job', 'marital', 'education', 'default', 'housing', 'loan', 'month'])])
````

````python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()


model = Pipeline([('transfomer', transform), ('modeling', LogisticRegression())])
````

````python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X, y , test_size = .2)
model.fit(x_train , y_train)
````
